```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib qt5

nX = 10                # number of grid points
domainWidth = 1e6      # meters
dX=domainWidth/nX
timeStep = 100         # years
nYears = 20000        # years
nSteps=(int)(nYears/timeStep)
flowParam = 1e4        # m horizontal / yr
snowFall = 0.5         # m / y
plotLimit=4000

elevation= np.zeros(nX+2)
flow= np.zeros(nX+1)

fig, ax =plt.subplots()
ax.set_ylim([0,plotLimit])
ax.plot( elevation)
plt.show( block=False )


for iTimes in range(0,nSteps):
    for ix in range(0,nX+1):
        flow[ix] = ( elevation[ix] - elevation[ix+1] ) / dX * flowParam  * \
                   ( elevation[ix]+elevation[ix+1] ) / 2 / dX
    for ix in range(1,nX+1):
        elevation[ix]=elevation[ix] + ( snowFall + flow[ix-1] - flow[ix] ) * timeStep
        
    print("Year",iTimes*timeStep)
    ax.clear()
    ax.plot( elevation)
    plt.show( block=False )
    fig.canvas.draw()
ax.clear()
ax.plot( elevation)
plt.show( block=False )
fig.canvas.draw()
```
